#### Problema 2:
Ecuación simple
Enunciado: $y - 5 = 12$

* Paso 1: Análisis y Resolución
* Planteamiento:

Tenemos una variable $y$ a la que se le resta 5.

* Ecuación: $y - 5 = 12$
* Despeje:
 Para dejar la $y$ sola, el 5 que está restando pasa al otro lado sumando.$y = 12 + 5$Solución: $y = 17$

```python
# Problema 2: Despeje de resta
resultado_derecho = 12
valor_restado = 5

# y = 12 + 5
y = resultado_derecho + valor_restado
print(f"El valor de y es: {y}")
```

#### Problema 4: Suma de variables
(m)Enunciado: $12 = m + 3m$
* Paso 1: Análisis y ResoluciónPlanteamiento: Tenemos una $m$ más tres $m$.
* Simplificación:
Sumamos las variables semejantes ($1 + 3 = 4$).$12 = 4m$
* Despeje: El 4 que multiplica a la $m$ pasa dividiendo al 12.$m = 12 / 4$Solución: $m = 3$

```python
# Problema 4: Agrupación de variables
total = 12
coeficiente_m = 1 + 3 # m + 3m = 4m

m = total / coeficiente_m
print(f"El valor de m es: {int(m)}")
```

#### Problema 6: Ecuación con resta de variables
* Enunciado:

$56 + 2b - 3b = 16$

$2b - 3b$ nos da $-1b$ o simplemente $-b$.$56 - b = 16$

$-b = 16 - 56$$-b = -40$
* Solución: Multiplicamos por $-1$ ambos lados para que la variable sea positiva.

#### Problema 8:
* Variables en ambos ladosEnunciado:

$7y - 5 = 3y + 7$
* Paso 1: Análisis y ResoluciónAgrupación:
Pasamos las $y$ a la izquierda y los números a la derecha.

$7y - 3y = 7 + 5$ Simplificación:$4y = 12$
* Despeje: El 4 pasa dividiendo.$y = 12 / 4$
* Solución: $y = 3$

```python
# Problema 8: Variables en ambos lados
# (7-3)y = 7 + 5 -> 4y = 12
y = (7 + 5) / (7 - 3)
print(f"El valor de y es: {int(y)}")
```
